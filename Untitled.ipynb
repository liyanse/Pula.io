````python
from scipy.optimize import newton

# Function to calculate the NPV for a given discount rate and future savings
def calculate_npv(discount_rate, initial_cost, future_savings):
    return -initial_cost + np.sum(future_savings / (1 + discount_rate) ** np.arange(len(future_savings)))

# Calculate IRR for Scenario 1
irr_scenario1 = newton(calculate_npv, 0.06, args=(initial_cost, future_savings_scenario1))

# Calculate IRR for Scenario 2
irr_scenario2 = newton(calculate_npv, 0.06, args=(initial_cost, future_savings_scenario2))

print("IRR for Scenario 1:", irr_scenario1)
print("IRR for Scenario 2:", irr_scenario2)
````
